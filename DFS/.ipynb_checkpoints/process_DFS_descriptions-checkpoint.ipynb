```python
import pandas as pd
import numpy as np
```

```python
def reformat_DFS(descriptions):
    reformatted = []
    for description in descriptions:
        conditions = description.split(' AND ')
        formatted_description = {}
        for condition in conditions:                    
            if '.isnull()' in condition:
                attribute = condition.split('.isnull()')[0]
                value = 'unknown'
                
            elif '>=' in condition:
                attribute = condition.split('>=')[0]
                value = str(condition.split('>=')[1])
                
            elif '<=' in condition:
                attribute = condition.split('<=')[0]
                value = str(condition.split('<=')[1])
                
            elif '<'  in condition:
                attribute, value = condition.split('<')
                attribute = attribute.split('.')[0]
                value = value.strip("'")
                
            elif '>'  in condition:
                attribute, value = condition.split('>')
                attribute = attribute.split('.')[0]
                value = value.strip("'")
                
            elif '[' in condition:
                attribute = condition.split('[')[0]
                value = str(condition.split('[')[1])
                attribute = attribute.split(':')[0]
                            
            else: 
                attribute, value = condition.split('==')
                attribute = attribute.split('.')[0]
                value = value.strip("'")
            
            if condition == "Age=='>=25.5'":
                attribute = 'Age'
                value = '>=25.5'
                
            elif condition == "Age=='<25.5'":
                attribute = 'Age'
                value = '<25.5'
            
            if condition == "Duration=='<15.5'":
                attribute = 'Duration'
                value = '<15.5'
                
            elif condition == "Duration=='>=15.5'":
                attribute = 'Duration'
                value = '>=15.5'
            
            if condition == "Credit amount=='<3913.5'":
                attribute = 'Credit amount'
                value = '<3913.5'
            elif condition == "Credit amount=='>=3913.5'":
                attribute = 'Credit amount'
                value = '>=3913.5'
   
            formatted_description[attribute] = value
        reformatted.append(formatted_description)
    return reformatted

DFS_df = pd.read_csv('unprocessed_DFS_descriptions.csv')
DFS_df = DFS_df[DFS_df['subgroup'] != 'Dataset']
DFS_df['description'] = reformat_DFS(list(DFS_df['subgroup']))
DFS_df['WRAcc score'] = DFS_df['quality']
DFS_df = DFS_df.drop(['subgroup', 'quality'], axis=1)
DFS_df
```

```python
bla = DFS_df[['description', 'WRAcc score']][:100]
```

```python
bla.to_csv('DFS_descriptions.csv')
```
