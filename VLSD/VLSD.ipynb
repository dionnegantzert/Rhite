## VLSD algorithm

VLSD (Vertical List Subgroup Discovery) [[1]](#1) is a subgroup discovery algorithm that combines an equivalence class exploration strategy and a pruning strategy based on optimistic estimate, and that is based on the vertical list data structure.

## Running the VLSD algorithm

To run the VLSD algorithm on a dataset, it is necessary to follow these steps:

- Load the dataset into a Pandas `DataFrame` object.
- Set the target, which must be a tuple of the form (column_name, value).
- Select the quality measure and optimistic estimate to use.
- Create the VLSD model with the desired parameters and run it.

The following is an example of running this algorithm on a dataset:

## Load Data and Preprocessing

```python
df = pd.read_excel("../german_credit_data.xlsx")
df['label'] = (df.Risk == 'good').astype(int)
df.drop(['Unnamed: 0','Risk', 'Sex'],axis=1,inplace=True)
df = df.replace(np.nan, 'unknown', regex=True)
df['Age'] = df['Age'].apply(lambda x: '>=25.5' if x >= 25.5 else '<25.5')
df['Credit amount'] = df['Credit amount'].apply(lambda x: '>=3913.5' if x >= 3913.5 else '<3913.5')
df['Duration'] = df['Duration'].apply(lambda x: '>=15.5' if x >= 15.5 else '<15.5')
df['Job'] = df['Job'].astype(str)
df['label'] = df['label'].astype(str)
```

```python
dataset = df
target = ('label', '1')
model = VLSD(quality_measure = WRAcc(), q_minimum_threshold  = -1, optimistic_estimate = WRAccOptimisticEstimate1(),
             oe_minimum_threshold = -1, sort_criterion_in_s1 = VLSD.SORT_CRITERION_QUALITY_DESCENDING, 
             sort_criterion_in_other_sizes = VLSD.SORT_CRITERION_QUALITY_DESCENDING, 
             vertical_lists_implementation = VLSD.VERTICAL_LISTS_WITH_BITSETS, 
             write_results_in_file = True, file_path = "./VLSD_descriptions.txt")
model.fit(dataset, target)
```

## Results

Running the following cell, we get the subgroups obtained by the algorithm:

```python
raw_data = []
with open("./VLSD_descriptions.txt", "r") as file:
    for current_line in file:
        raw_data.append(str(current_line.strip()))
```

```python
def parse_data(raw_data):
    parsed_data = []
    for entry in raw_data:
        description_match = re.search(r"Description: \[(.*?)\]", entry)
        target_match = re.search(r"Target: label = '(\d)'", entry)
        tp_match = re.search(r"Sequence of instances tp = bitarray\('(.*?)'\)", entry)
        fp_match = re.search(r"Sequence of instances fp = bitarray\('(.*?)'\)", entry)
        wracc_match = re.search(r"Quality Measure WRAcc = ([\d\.]+)", entry)
        optimistic_estimate_match = re.search(r"Optimistic Estimate WRAccOptimisticEstimate1 = ([\d\.]+)", entry)
        tp_val_match = re.search(r"tp = (\d+)", entry)
        fp_val_match = re.search(r"fp = (\d+)", entry)
        TP_val_match = re.search(r"TP = (\d+)", entry)
        FP_val_match = re.search(r"FP = (\d+)", entry)

        parsed_data.append({
            "description": description_match.group(1) if description_match else None,
#             "Target": int(target_match.group(1)) if target_match else None,
#             "Sequence of instances tp": tp_match.group(1) if tp_match else None,
#             "Sequence of instances fp": fp_match.group(1) if fp_match else None,
            "WRAcc score": float(wracc_match.group(1)) if wracc_match else 0,
            "WRAcc optimistic": float(optimistic_estimate_match.group(1)) if optimistic_estimate_match else 0,
            "tp": int(tp_val_match.group(1)) if tp_val_match else None,
            "fp": int(fp_val_match.group(1)) if fp_val_match else None,
            "TP": int(TP_val_match.group(1)) if TP_val_match else None,
            "FP": int(FP_val_match.group(1)) if FP_val_match else None
        })

    return parsed_data
```

```python
parsed_data = parse_data(raw_data)
df = pd.DataFrame(parsed_data)
df = df.sort_values(by='WRAcc score', ascending=False)[:100]
```

```python
def reformat_VLSD(descriptions):
    reformatted = []
    for description in descriptions:
        rule = description.split(', ')
        result_dict = dict()

        for r in rule:
            # Split each pair by the equals sign
            key, value = r.split(' = ')
            result_dict[key] = value.strip("'")
        reformatted.append(result_dict)
    return reformatted


def reformat_DVLSD(descriptions):
    reformatted = []
    for description in descriptions:
        conditions = description.split(' AND ')
        formatted_description = {}
        for condition in conditions:                    
            if '.isnull()' in condition:
                attribute = condition.split('.isnull()')[0]
                value = 'unknown'
                
            elif '>=' in condition:
                attribute = condition.split('>=')[0]
                value = str(condition.split('>=')[1])
                
            elif '<=' in condition:
                attribute = condition.split('<=')[0]
                value = str(condition.split('<=')[1])
                
            elif '<'  in condition:
                attribute, value = condition.split('<')
                attribute = attribute.split('.')[0]
                value = value.strip("'")
                
            elif '>'  in condition:
                attribute, value = condition.split('>')
                attribute = attribute.split('.')[0]
                value = value.strip("'")
                
            elif '[' in condition:
                attribute = condition.split('[')[0]
                value = str(condition.split('[')[1])
                attribute = attribute.split(':')[0]
                            
            else: 
                attribute, value = condition.split('==')
                attribute = attribute.split('.')[0]
                value = value.strip("'")
            
            if condition == "Age=='>=25.5'":
                attribute = 'Age'
                value = '>=25.5'
                
            elif condition == "Age=='<25.5'":
                attribute = 'Age'
                value = '<25.5'
            
            if condition == "Duration=='<15.5'":
                attribute = 'Duration'
                value = '<15.5'
                
            elif condition == "Duration=='>=15.5'":
                attribute = 'Duration'
                value = '>=15.5'
            
            if condition == "Credit amount=='<3913.5'":
                attribute = 'Credit amount'
                value = '<3913.5'
            elif condition == "Credit amount=='>=3913.5'":
                attribute = 'Credit amount'
                value = '>=3913.5'
   
            formatted_description[attribute] = value
        reformatted.append(formatted_description)
    return reformatted
```

```python
df['description'] = reformat_VLSD(list(df['description']))
```
