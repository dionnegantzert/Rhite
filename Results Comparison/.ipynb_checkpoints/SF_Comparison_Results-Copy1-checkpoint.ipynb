# Compare Results of Clustering, DFS and VLSD

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
clustering_df = pd.read_csv('kmedoids_lr_results.csv')
DFS_df = pd.read_csv('DFS_lr_results.csv')
VLSD_df = pd.read_csv('VLSD_lr_results.csv')

clustering_df = clustering_df.drop_duplicates()
DFS_df = DFS_df.drop_duplicates()
VLSD_df = VLSD_df.drop_duplicates()

clustering_df = clustering_df[clustering_df['description'] != '{}']
DFS_df = DFS_df[DFS_df['description'] != '{}']
VLSD_df = VLSD_df[VLSD_df['description'] != '{}']

clustering_df['model'] = 'Kmedoids'
VLSD_df['model'] = 'VLSD'
DFS_df['model'] = 'DFS'

colors = ['#ff6d6d', '#c2891c', '#5d9a2e']
palette = { 'Kmedoids': colors[0], 'VLSD': colors[1], 'DFS': colors[2] }

combined_df = pd.concat([clustering_df, VLSD_df, DFS_df])
```

# Internal Subgroup Fairness

### Demographic Parity Difference

```python
plt.figure(figsize=(15, 6))
for model in combined_df['model'].unique():
    subset = combined_df[combined_df['model'] == model]
    sns.kdeplot(subset['demographic_parity_difference'], label=model, color=palette[model], fill=True)

# Add a vertical line at x = 0
plt.axvline(x=0, color='black', linestyle='--')

# Add labels and title
plt.xlabel('Demographic Parity Difference Score')
plt.ylabel('Density')
plt.title('Density Plot of DPD Scores by Model')
plt.legend(title='Model')

# Show the plot
plt.show()
```

```python
def line_plot(feature, max_x=30):
    
    # y1 = np.abs(clustering_df['statistical_parity_subgroup'][:max_x])
    # y2 = np.abs(VLSD_df['statistical_parity_subgroup'][:max_x])
    # y3 = np.abs(DFS_df['statistical_parity_subgroup'][:max_x])

    y1 = clustering_df[feature][:max_x]
    y2 = VLSD_df[feature][:max_x]
    y3 = DFS_df[feature][:max_x]
    # y4 = data_clustering['statistical_parity_subgroup'][:max_x]

    # y1_sorted = np.abs(clustering_df['statistical_parity_subgroup'][:max_x].sort_values(ascending=True))
    # y2_sorted = np.abs(VLSD_df['statistical_parity_subgroup'][:max_x].sort_values(ascending=True))
    # y3_sorted = np.abs(DFS_df['statistical_parity_subgroup'][:max_x].sort_values(ascending=True))

    y1_sorted = clustering_df[feature][:max_x].sort_values(ascending=True)
    y2_sorted = VLSD_df[feature][:max_x].sort_values(ascending=True)
    y3_sorted = DFS_df[feature][:max_x].sort_values(ascending=True)
    # y4_sorted = data_clustering['statistical_parity_subgroup'][:max_x].sort_values(ascending=True)


    # plt.axhline(y=np.mean(y2), color='red', linestyle='--')

    x = np.arange(max_x)  # x values

    fig, ax = plt.subplots(figsize=(15,10))
    ax.plot(x, list(y1_sorted), label='Kmedoids')
    ax.plot(x, list(y2_sorted), label='VLSD')
    ax.plot(x, list(y3_sorted), label='DFS')
    # ax.plot(x, list(y4_sorted), label='Data')

    ax.axhline(y=np.mean(y1), color='blue', linestyle='--')
    ax.axhline(y=np.mean(y2), color='orange', linestyle='--')
    ax.axhline(y=np.mean(y3), color='green', linestyle='--')
    # ax.axhline(y=np.mean(y4), color='red', linestyle='--')

    # Add labels, title, and legend
    ax.set_xlabel('Subgroups')
    ax.set_ylabel(feature)
    ax.set_title(feature + 'for each method')
    ax.legend()

    # Show the plot
    plt.show()
line_plot('demographic_parity_difference')
```

### Equalized Odds Difference

```python
plt.figure(figsize=(15, 6))
for model in combined_df['model'].unique():
    subset = combined_df[combined_df['model'] == model]
    sns.kdeplot(subset['equalized_odds_difference'], label=model, color=palette[model], fill=True)

# Add a vertical line at x = 0
plt.axvline(x=0, color='black', linestyle='--')

# Add labels and title
plt.xlabel('Equalized Odds Difference Score')
plt.ylabel('Density')
plt.title('Density Plot of EOD Scores by Model')
plt.legend(title='Model')

# Show the plot
plt.show()
```
